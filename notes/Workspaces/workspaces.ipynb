### Workspace:

- Inspired by the Cargo concept of the same name, a workspace is "a collection of one or more packages, called workspace members, that are managed together.

#### To create a workspace:

- add a ```tool.uv.workspace``` table to a ```pyproject.toml```, which will implicitly create a workspace rooted at that package

Note:
- running uv init will add newly created workspace 

- defining a workspace, you must specify the members (required) and exclude (optional) keys, which direct the workspace to include or exclude specific directories as members respectively, and accept lists of globs in ```pyproject.toml``` file
- workspace members can be either applications or libraries
- every workspace needs a root, which is also a workspace member.
