### Publishing Package:

```uv publish```

Note:
- Set a PyPI token with --token or UV_PUBLISH_TOKEN
- publishing to PyPI from GitHub Actions or another Trusted Publisher, you don't need to set any credentials
- To add trusted publisher: https://docs.pypi.org/trusted-publishers/adding-a-publisher/


- When publishing a package, we recommend running uv build --no-sources to ensure that the package builds correctly when tool.uv.sources is disabled, as is the case when using other build tools, like pypa/build.

- If you're using a custom index through ```[[tool.uv.index]]```, add ```publish-url``` and use ```uv publish --index <name>```
    - ```uv publish --index <name>```, the pyproject.toml must be present
