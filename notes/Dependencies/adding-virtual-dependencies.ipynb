### Install virtual dependencies:

- "virtual": dependency itself is not installed as a package, but its dependencies are
- to treat a dependency as virtual show below,
    - ```package = false``` on source

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/virtual.png')
new_image = img.resize((430, 150))
display(new_image)
````

- if a dependency sets ```tool.uv.package = false```, it can be overridden

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/virtual1.png')
new_image = img.resize((410, 150))
display(new_image)
````

#### virtual dependency with workspace:

Options to build virtual dependencies:
1. ```tool.uv.package = false``` flag in workspace can be used to explicitly make dependency virtual
2. shown below,

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/parent.png')
new_image = img.resize((180, 190))
display(new_image)
````

- if the child project package has no build system in ```pyproject.toml``` file, then the child package will not be installed instead its transitive dependency such as ```anyio``` is installed. 

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/child.png')
new_image = img.resize((220, 140))
display(new_image)
````

In contrast, if parent declared as dependency on child like below, then the child package would be built and installed along with transitive dependencies

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/parent2.png')
new_image = img.resize((240, 240))
display(new_image)
````
