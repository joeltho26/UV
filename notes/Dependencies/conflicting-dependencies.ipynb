### Conflicting Dependencies:

Link: https://docs.astral.sh/uv/concepts/resolution

- ```uv``` resolves all project dependencies together, including optional dependencies ("extras") and dependency groups
- if dependencies declared in one section are not compatible with those in another section, ```uv``` will fail to resolve the requirements of the project with an error as below
- eg: to declare that the ```optional-dependency groups``` extra1 and extra2 are incompatible

example, consider two sets of optional dependencies that conflict with one another as shown below

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/conflict1.png')
new_image = img.resize((290, 120))
display(new_image)
````

- now, when we run ```uv lock``` will raise an error

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/conflict.png')
new_image = img.resize((230, 190))
display(new_image)
````

- now, running uv lock will succeed. However, now you cannot install both extra1 and extra2 at the same time

    ```uv sync --extra extra1 --extra extra2```

    - running the above command, an error occurs because installing both extra1 and extra2 would result in installing two different versions of a package into the same environment
