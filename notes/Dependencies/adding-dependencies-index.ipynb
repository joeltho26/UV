### Install dependencies via source indexes:

```uv add torch --index pytorch=https://download.pytorch.org/whl/cpu```


````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/index.png')
new_image = img.resize((390, 250))
display(new_image)
````

- using an index source pins a package to the given index  it will not be downloaded from other indexes
- when defining an index, an explicit flag can be included to indicate that the index should only be used for packages that explicitly specify it in tool.uv.sources
- if explicit is not set, other packages may be resolved from the index, if not found elsewhere
