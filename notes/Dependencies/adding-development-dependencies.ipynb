### Installing development dependencies:

```uv add --dev pytest```

- development dependencies are local only and will not be included in the project requirements when published to PYPI (python package indexes)
- development dependencies can have entries in ```tool.uv.sources```

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/dev.png')
new_image = img.resize((190, 130))
display(new_image)
````

- development dependency flags, 
    - ```--dev```

        - inclusion of ```dev``` dependencies along with other project dependencies

    - ```--only-dev```

        - inclusion of ```dev``` dependencies only and no other project dependencies
    
    - ```--no-dev```

        - exclusion of ```dev``` dependencies

Note:
- ```dev``` group is synced by default

#### Dependency groups:

- Development Dependencies can be divided into multiple groups

```uv add --group lint ruff```

- Dependency groups flags:

    - ```--all-groups```
        
        - include all groups (development dependency groups)

    - ```--no-default-groups```

        - exclude all groups (development dependency groups)
    
    - ```--groups```

        - include groups (development dependency groups) along with project dependencies

    - ```--only-group```
    
        - include groups (development dependency groups) only and not other project dependencies

    - ```--no-group```

        - exclude specific development dependency groups mentioned
