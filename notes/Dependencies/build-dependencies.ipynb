### Build dependencies:

- there are certain dependencies declared that are used to build project but not required to run it. 
- these dependencies are specified in the ```[build-system]``` table under ```build-system.requires```


Note:
- when publishing a package, we recommend running ```uv build --no-sources``` to ensure that the package builds correctly when ```tool.uv.sources``` is disabled, as is the case when using other build tools, like pypa/build
