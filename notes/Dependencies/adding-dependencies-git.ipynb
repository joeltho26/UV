- https:

    ```uv add git+https://github.com/psf/requests```


    ```uv add “httpx @ git+https://github.com/encode/httpx”``` (install dependencies from github)

- ssh:

    ```uv add git+ssh://git@github.com/encode/httpx``` (add git dependencies via ssh)
