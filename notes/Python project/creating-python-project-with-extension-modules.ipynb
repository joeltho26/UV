### Creating python project using extension modules:

- most python projects are "pure Python", meaning they do not define modules in other languages like C, C++, FORTRAN, or Rust
- however, projects with extension modules are often used for performance sensitive code.

Creating a project with an extension module requires choosing an alternative build system:
- maturin (RUST projects)
- scikit-build-core (C, C++, FORTRAN, Cython)

maturin (RUST projects)

```uv init --build-backend maturin example-ext```

````python
"""
example-ext
├── .gitignore
├── .python-version
├── Cargo.toml (RUST package installer)
├── README.md
└── pyproject.toml 
└── src
    ├── lib.rs
    └── example_ext
        └── __init__.py
        └── _core.pyi
"""
````

- ```src/lib.rs``` file contains the code shown below

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/librs.png')
new_image = img.resize((510, 300))
display(new_image)
````

- ```__init__.py``` file contains the below code

````python
from example_ext._core import hello_from_bin


def main() -> None:
    print(hello_from_bin())
````

```cd example-ext```

```uv run example-ext```


- Output:
    - "Hello from example-ext"
