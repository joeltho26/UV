### Create a python project application using uv:
- while trying to create a project, we can create it for <b>application</b> or <b>libraries</b>
- by default, we can create for application or also use ```--app``` flag
- by using ```--lib``` flag, we can create project for library instead

- the below command will create a python project application folder inside the folder "/example-app"

    ```uv init example-app```

- project files created below after running the above command:

````python
"""
example-app
├── .gitignore
├── .python-version
├── README.md
├── main.py
└── pyproject.toml 
"""
````

- creating the project will create project.toml file which contains the metadata of the project

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/output.png')
new_image = img.resize((370, 200))
display(new_image)
````

Directory:
- the project will be created inside the directory mentioned

    ```uv init myproject```

- we can also create project inside the directory by executing the below commands,

    ```mkdir myproject```

    ```cd myproject```

    ```uv init```

Note:
- if there is a .toml already present inside the folder where we are trying to create the new project will raise an error



Running program files inside the project folder
- after creating the project then we can run the main.py file (consist of logic) as shown below,

    ```cd example-app```
    
    ```uv run main.py```

````python
def main():
    print("Hello from example-app!")


if __name__ == "__main__":
    main()
````
