### Create a Python project library using uv:

-  provides functions and objects for other projects to consume
- they are intended to be built and distributed

```uv init --lib example-lib```

````python
"""
example-lib
├── .gitignore
├── .python-version
├── README.md
└── pyproject.toml 
└── src
    └── example-lib
        └── py.typed
        └── __init__.py
"""
````

````python
from PIL import Image 
from IPython.display import display

img = Image.open('images/lib.png')
new_image = img.resize((370, 300))
display(new_image)
````

- add the below function inside the ```__init__.py``` file

````python
def hello() -> str:
    return "Hello from example-lib!"
````

- now to import and execute the library function,

    ```cd example-lib```
    
    ```uv run python -c "import example_lib; print(example_lib.hello())"```
